```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.utils import to_categorical
```

```python
# 1. Load the MNIST dataset (digits 0-9)
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
# 2. Normalize the data (convert from 0-255 to 0-1)
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0
```

```python
# 3. Reshape data to fit CNN input (samples, height, width, channels)
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)
```

```python
# 4. Convert labels to categorical (one-hot encoding)
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)
```

```python

# 5. Build the CNN model
model = Sequential([
    Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')  # 10 classes for digits 0-9
])
```

```python
# 6. Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# 7. Train the model
model.fit(X_train, y_train_cat, epochs=5, batch_size=64, validation_data=(X_test, y_test_cat))
```

```python
# 8. Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test_cat)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

```python
# 9. Test with a single image from test set
plt.imshow(X_test[0].reshape(28, 28), cmap='gray')
plt.title("Test Image Example")
plt.show()
```

```python
test_image = X_test[0].reshape(1, 28, 28, 1)
prediction = model.predict(test_image)
predicted_digit = np.argmax(prediction)
print("Predicted digit:", predicted_digit)
```

```python
# 10. Save the model (optional)
model.save("digit_recognition_model.h5")
```
